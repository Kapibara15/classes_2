```python
full_country_stats = oecd_bli.merge(
    gdp_per_capita_2020,
    left_on="Country",
    right_on="Entity",
    how="inner"
)
full_country_stats
```

```python
full_country_stats = full_country_stats[["Country", "Life satisfaction", "GDP per capita (USD)"]]
full_country_stats = full_country_stats.sort_values(by = "GDP per capita (USD)")
full_country_stats
```

```python
full_country_stats.to_csv(
    datapath / "Liveset_full.csv",
    index=False
)
```

```python
min_gdp= 23_500
max_gdp=62_500
```

```python
country_stats = full_country_stats[
    (full_country_stats["GDP per capita (USD)"] >= min_gdp)
    & (full_country_stats["GDP per capita (USD)"] <= max_gdp)
]
country_stats.to_csv(
    datapath / "Liveset.csv",
    index=False
)
```

```python
lifeset = pd.read_csv("data/Liveset/Liveset.csv")

#przygotowwujemy dane
lifeset.rename(
    columns= {"GDP per capita (USD)": "PKB na mieszkanca (USD)"},
    inplace=True
)
lifeset.rename(
    columns= {"Life satisfaction": "Satysfakcja z zycia"},
    inplace=True)
lifeset
```

```python
#wizualizacja danych na wykresie punktowym
lifeset.plot(kind="scatter",grid=True,
             x="PKB na mieszkanca (USD)",y="Satysfakcja z zycia")
plt.show()
```

```python
#Tworzenie modelu
X = lifeset[["PKB na mieszkanca (USD)"]]
y = lifeset[["Satysfakcja z zycia"]]

#wybieram model liniowy
model = LinearRegression()

#trenduje model
model.fit(X,y)

#obliczam wspÃ³Å‚czynniki modelu
theta_0=model.intercept_[0]  #aby wydobyÄ‡ pierwszy element z macierzy
theta_1=model.coef_[0][0]
print(f"theta_0: {theta_0}")
print(f"theta_1: {theta_1}")
```

```python
#kolejny wykres
lifeset.plot(kind="scatter",grid=True,
             x="PKB na mieszkanca (USD)",y="Satysfakcja z zycia")
plt.plot(X, theta_0+theta_1*X)
plt.text(45000, 6.5, f"$\ttheta_0 = {theta_0.round(2)}$",
         color = "blue")
plt.text(45000, 6.3, f"$\ttheta_1 = {theta_1 * 1e5:.2f}\cdot 10^{{-5}}$",
         color = "blue")
plt.show()
```

```python
# PKB dla Cypru
cyprus_gdp_per_capita_2020= (
    gdp_per_capita_2020[gdp_per_capita_2020["Entity"]=="Cyprus"][["GDP per capita (USD)"]]
)
cyprus_gdp_per_capita_2020.rename(
    columns= {"GDP per capita (USD)": "PKB na mieszkanca (USD)"},
    inplace=True
)
cyprus_gdp_per_capita_2020
```

```python
cyprus_predicted_life_satisfaction= model.predict(cyprus_gdp_per_capita_2020)
cyprus_predicted_life_satisfaction
```

```python
from sklearn.neighbors import KNeighborsRegressor
model_kmn = KNeighborsRegressor(n_neighbors=3)

#trendujemy model
model_kmn.fit(X,y)

#uzyskuje przewidywania dla Cypru
model_kmn.predict(cyprus_gdp_per_capita_2020)
```
